```python
import requests

#AlphaVantage API key
api_key = '8U8OGK31EO6CESRW'

#symbol take the stock which we want the data for
symbol = input("Enter the stock symbol: ")
```

```python

#request to fetch data
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol={symbol}&apikey={api_key}'

response = requests.get(url)
if response.status_code == 200:
    data = response.json()

    time_series = data['Time Series (Daily)']
    
    for date, info in time_series.items():
        # Access the adjusted close price for each day
        close_price = info['5. adjusted close']
        
        # Print the date and adjusted close price
        print(f"Date: {date}, Adjusted Close Price: {close_price}")

else:
    print(f"Error occurred. Status code: {response.status_code}")
```

Testing Linear regression with only time series data

```python

from sklearn.linear_model import LinearRegression
# Parsing and preparing the data for prediction
time_series = data['Time Series (Daily)']
dates = sorted(time_series.keys())
closing_prices = [float(time_series[date]['4. close']) for date in dates]

# Creating X and y for linear regression
X = [[i] for i in range(1, len(dates) + 1)]
y = closing_prices

# Fitting the linear regression model
regressor = LinearRegression()
regressor.fit(X, y)

# Predicting the next closing price
next_day = len(dates) + 1
next_price = regressor.predict([[next_day]])[0]

print(f"Predicted closing price for the next day: ${next_price:.2f}")
```
